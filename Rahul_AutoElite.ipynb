```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
budget = pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE/refs/heads/main/AutoElite_Budget.csv")
budget.head()
```

```python
transaction = pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE/refs/heads/main/AutoElite_Actual.csv")
transaction.head()
```

```python
budget.columns
```

```python
transaction.columns
```

```python
budget.columns
budget.columns = budget.columns.str.lower().str.replace(" ","_")
budget.columns
```

```python
transaction.columns
transaction.columns = transaction.columns.str.lower().str.replace(" ","_")
transaction.columns
```

```python
actual = transaction.groupby(['month','region','state','model'])['quantity_sold'].sum().reset_index()
actual
```

```python
merged_table = pd.merge(budget, actual, on=['month','region','state','model'], how='left')
print(merged_table.head())
```

```python
merged_table['variance'] = merged_table['quantity_sold']-merged_table['budgeted_quantity']
merged_table
#merged_table.head()
```

```python
month_mapping = {'January':1,
                 'February':2,
                 'March':3,
                 'April':4,
                 'May':5,
                 'June':6,
                 'July':7,
                 'August':8,
                 'September':9,
                 'October':10,
                 'November':11,
                 'December':12}
merged_table['month_number'] = merged_table['month'].map(month_mapping)
merged_table
```

```python
monthly_variance = merged_table.groupby(['month','month_number'])['variance'].sum().reset_index()
sorted_monthly_variance = monthly_variance.sort_values('month_number')
sorted_monthly_variance
```

```python
plt.bar(x=sorted_monthly_variance['month'],height= sorted_monthly_variance['variance'])
plt.xlabel('month')
plt.ylabel('variance')
plt.title("Monthly Sales variance (in unit sold)")
plt.xticks(ticks=sorted_monthly_variance['month'],rotation = 90)
plt.show()
```

```python
regionwise_variance = merged_table.groupby(['region'])['variance'].sum().reset_index()
regionwise_variance
```

```python
plt.bar(x= regionwise_variance['region'],height= regionwise_variance['variance'])
plt.show()
```

```python
plt.bar(x= regionwise_variance['region'], height= regionwise_variance['variance'])
plt.xlabel('region')
plt.ylabel('variance')
plt.title('regionwise Sales Variance (in Unit Sold)')
plt.show()
```

```python
statewise_variance = merged_table.groupby(['state'])['variance'].sum().reset_index()
statewise_variance
```

```python
plt.bar(x= statewise_variance['state'],height= statewise_variance['variance'])
plt.xlabel('state')
plt.ylabel('variance')
plt.title('statewise variance (in Unit sold)')
plt.xticks(ticks=statewise_variance['state'], rotation = 90)
plt.show()
```
